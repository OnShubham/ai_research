```python
from langchain_community.document_loaders.pdf import PyPDFLoader
import pprint

file_path = "../data/Learning_Python.pdf"
loader = PyPDFLoader(file_path)


docs = loader.load()
docs[0]

pprint.pp(docs[0].metadata)
```

```python
from langchain_text_splitters import  RecursiveCharacterTextSplitter

text_spilter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", ".", " ", ""]

    
)

chunk = text_spilter.split_documents(docs)
print(len(chunk))
# print(chunk[500].page_content)


# pprint.pp(chunk[0])
```

```python
# embbeding

from sentence_transformers import SentenceTransformer


model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

texts = [doc.page_content for doc in chunk]

embbeding = model.encode(texts)
print(embbeding.shape)

```

```python
from langchain.vectorstores import FAISS

vectorstore = FAISS.from_documents(
    chunk,
    embedding
)

vectorstore.save_local("faiss_index")
```
